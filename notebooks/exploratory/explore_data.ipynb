```python
import psycopg2
import pandas as pd
import geopandas as gp
```

```python
DBNAME = "opportunity_youth"
```

```python
conn = psycopg2.connect(dbname=DBNAME)
```

```python
pd.read_sql("SELECT * FROM pums_2017 LIMIT 100;", conn)
```

```python
#get table schema using postgresql/psycopg2 syntax
QUERY = """
SELECT table_schema, table_name
FROM information_schema.tables
WHERE (table_schema = 'public')
ORDER BY table_schema, table_name;
"""

table_info = pd.read_sql(QUERY, conn)
table_names = table_info['table_name']
table_names
for table in table_names:
    print(table)
```

```python
#find table contents/keys 
for table in table_names:
    QUERY = f"""
        SELECT *
        FROM  {table}
        WHERE false
        """
    
    print("".join(['\033[1m', f'{table}', '\033[0m']))#make column name bold
    print(pd.read_sql(QUERY, conn))
```

```python
#find table columns/keys of pums_2017

QUERY = """
    SELECT *
    FROM pums_2017
    WHERE false
    """
    
print("".join(['\033[1m', 'pums_2017', '\033[0m']))#make column name bold
print(pd.read_sql(QUERY, conn))
```

```python
#explore table columns/keys of pums_2017

QUERY = """
    SELECT *
    FROM pums_2017 p
    WHERE p.rt = 'p'
    """
    
print("".join(['\033[1m', 'pums_2017', '\033[0m']))#make column name bold
print(pd.read_sql(QUERY, conn))
```

```python
#explore table columns/keys of puma_names_2010
table = 'puma_names_2010'

QUERY = f"""
    SELECT *
    FROM {table} p
    WHERE p.state_fips LIKE '053'
    """
    
print("".join(['\033[1m', f'{table}', '\033[0m']))#make column name bold
print(pd.read_sql(QUERY, conn))
```

```python
QUERY = """
    SELECT *
    FROM pums_2017 ps
    JOIN puma_names_2010 pn
    ON ps.puma = pn.puma
    WHERE pn.puma SIMILAR TO '1161(0|1|2|3|4|5)'
    AND ps.agep >= 16
    AND ps.agep <= 24
    AND ps.sch = '1'
    AND ps.dis = '2'
    AND ps.esr SIMILAR TO '%(3|6)%'
    ORDER BY pn
    """
db_south = pd.read_sql(QUERY, conn)
db_south
```

```python
south_king_county_db = pd.read_sql('''
        SELECT *
        FROM pums_2017 PS
        WHERE PS.puma SIMILAR TO '1161(0|1|2|3|4|5)'
        AND PS.agep >= 16
        AND PS.agep <= 24
        AND PS.sch = '1'
        AND PS.esr SIMILAR TO '%(3|6)%'
        
''', conn)
sum(south_king_county_db['pwgtp']) #total number of OY as per census data
```
