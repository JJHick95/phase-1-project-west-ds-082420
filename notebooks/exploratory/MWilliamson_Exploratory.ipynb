<B> establish connection with database </B>

```python
import psycopg2
import pandas as pd
DBNAME = "opportunity_youth"
conn = psycopg2.connect(dbname=DBNAME)
```

<b> get table names

```python
pd.read_sql('''SELECT table_name
                FROM information_schema.tables
                WHERE table_schema = 'public';
            ''', conn)
```

<b> check the head of each table

```python
#pums_2017 head
#var name descriptions found in pums_data_dictionary_2017.pdf
#esr = employment status recode
#agep = age
#sch = school enrollment
#jwmnp = travel time to work in minutes
#dis = disability

pd.read_sql('''
        SELECT esr
        FROM pums_2017
        LIMIT 10
''', conn)
```

```python
#pums_2017 column names
column_names = pd.read_sql('''
SELECT column_name, data_type
FROM INFORMATION_SCHEMA.COLUMNS
WHERE TABLE_NAME = 'pums_2017'
ORDER BY ordinal_position
''', conn)
pd.set_option('display.max_rows', column_names.shape[0] + 1)
column_names
```

<b> return a database pums2017 filtered for OY in SKC

```python
south_king_county_db = pd.read_sql('''
        SELECT *
        FROM pums_2017 PS
        WHERE PS.puma SIMILAR TO '1161(0|1|2|3|4|5)'
        AND PS.agep >= 16
        AND PS.agep <= 24
        AND PS.sch = '1'
        AND PS.esr SIMILAR TO '%(3|6)%'
        
''', conn)
sum(south_king_county_db['pwgtp'])
```
